### We can do a lot with plain text files and metadata

```python
import pandas as pd
```

```python
# load metadata.csv to pandas dataframe
df = pd.read_csv('gutenberg_selections/metadata.csv')
df
```

### Each row has a gutenberg id, and most documents have a corresponding [id].txt file in the full_text folder

```python
#os.walk reada and displays all the files in a location
import os

txtfilenames =[]
for a,b,c in os.walk("gutenberg_selections/full_text"):
    for name in c:
        if name.endswith(".txt"):
            txtfilenames.append(name)
txtfilenames[:10]
```

### We can use the dataframe as a reference to load specific files (like all work after 1850)

```python
df_post_1850 = df[df['date'] > 1850]
df_post_1850
```

### And then load the corresponding txt files

```python
myfiles = []
no_file = []

#loop gutenberg ids
for i in df_post_1850["gutenberg_id"]:
    filename = "gutenberg_selections/full_text/"+i+".txt"
    #the try statement is here in case any text file isn't found
    try:
        with open(filename) as f:
            txt = f.read()
        myfiles.append(txt)
    except:
        no_file.append(i)
```

### Because of how I've written the code, the txt files will now be stored in mylist in the order they appear in  df_post_1850

```python
len(myfiles)
```

```python
for i in no_file:
    df_post_1850 = df_post_1850[df_post_1850['gutenberg_id'] != i] 
df_post_1850
```

### I can tokenize the items in mylist like any list of strings

```python
token_list = []
for i in myfiles:
    #remember, this is the fast way, not the best way, to tokenize
    tokens = i.replace("\n", "").replace("\t", "").lower().split()
    token_list.append(tokens) 
```

### And now we can plot something simple like the densities of words ending in "ly" by date

```python
ly_densities = []
for t in token_list:
    total_length = len(t)
    lenght_of_ly = len([i for i in t if i.endswith('ly')])
    ly_densities.append(lenght_of_ly*1.0/total_length)
```

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

dates = df_post_1850['date']

plt.scatter(dates, ly_densities, alpha=.5)
plt.show()
```

### XML files use tags to mark up hierarchical data (often used for text data)

```python
#texts from http://webapp1.dlib.indiana.edu/vwwp
import xml.etree.ElementTree
xml_file_name = 'xml/VAB7046.xml'
e = xml.etree.ElementTree.parse(xml_file_name).getroot()
```

```python
from xml.dom import minidom
xmldoc = minidom.parse(xml_file_name)
itemlist = xmldoc.getElementsByTagName('div')
paragraphs = []
for div in itemlist:    
    if div.attributes['type'].value == "chapter":
        chapter_ps =  div.getElementsByTagName('p')
        paragraphs.extend(chapter_ps)
```

```python
#Note: you may have to install this library
from bs4 import BeautifulSoup
```

```python
xml_file2 = 'xml/VAB1828.xml'

with open(xml_file2) as f:
    xml2 = f.read()

bs = BeautifulSoup(xml2)
```

```python
jclifford = bs.find_all("said", {"who":"#jclifford"})
belliot = bs.find_all("said", {"who":"#belliot"})
```

### To tokenize each speaker's dialogue, we need to convert lists of xml tags to strings of only text data, but that requires only one more line of code

```python
token_list_jc_be = []
for i in [jclifford, belliot]:
    dialogue = " ".join([x.text for x in i])
    #remember, this is the fast way, not the best way, to tokenize
    tokens = dialogue.replace("\n", "").replace("\t", "").lower().split()
    token_list_jc_be.append(tokens) 
```

```python
ly_densities = []
for t in token_list_jc_be:
    total_length = len(t)
    lenght_of_ly = len([i for i in t if i.endswith('ly')])
    ly_densities.append(lenght_of_ly*1.0/total_length)
```

### We can now plot the "ly" desnities for each character

```python
speakers = ["Joe Clifford", "Birdie Elliot"]
plt.scatter(speakers, ly_densities, alpha=.5)
plt.show()
```
